1. Read table into a SparkDataFrame

```python
df = spark.table("nemsis_subset") 
df.printSchema()
df.limit(5).display()
```

Basic EDA

```python
df.count()
from pyspark.sql import functions as F

(df.select([F.count(F.when(F.col(c).isNull(), 1)).alias(c) for c in df.columns])
   .display())
```

```python
from pyspark.sql import functions as F

df_feat = (df
   .withColumn("ts", F.to_timestamp("incident_timestamp", "yyyy-MM-dd HH:mm:ss"))
   .withColumn("date", F.to_date("ts"))
   .withColumn("hour", F.hour("ts"))
   .withColumn("dow", F.date_format("ts", "u").cast("int")))

df_feat.limit(5).display()
```
